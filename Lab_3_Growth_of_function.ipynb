# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
```

```python
import math
import time
import matplotlib.pyplot as plt
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0

    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return steps
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps

n_values = []
time_values = []


for i in range(1, 15):
    n = 2**i
    n_values.append(n)
    arr = list(range(n))

    actual_steps = binary_search(arr, n - 1)
    time_values.append(actual_steps)


plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='b', label='Observed Steps')


log_n_theory = [math.log2(n) + 1 for n in n_values]
plt.plot(n_values, log_n_theory, linestyle='--', color='r', label='Theoretical O(log n)')

plt.xlabel("Input size n")
plt.ylabel("Number of Operations (steps)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search)")
plt.xscale('log', base=2)
plt.yscale('log', base=2)
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend()
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
import matplotlib.pyplot as plt
import math
import time

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):
    merged = []
    left_idx = 0
    right_idx = 0

    while left_idx < len(left) and right_idx < len(right):
        if left[left_idx] < right[right_idx]:
            merged.append(left[left_idx])
            left_idx += 1
        else:
            merged.append(right[right_idx])
            right_idx += 1

    while left_idx < len(left):
        merged.append(left[left_idx])
        left_idx += 1

    while right_idx < len(right):
        merged.append(right[right_idx])
        right_idx += 1
    return merged

n_values = []
time_values = []
for i in range(1, 10):
    n = 100 * (2**i)
    n_values.append(n)

    data = list(range(n, 0, -1))

    start_time = time.time()
    merge_sort(data)
    end_time = time.time()

    time_values.append((end_time - start_time) * 1000)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='b', label='Observed Time (ms)')

if len(n_values) > 0 and n_values[-1] > 1:
    scaling_factor = time_values[-1] / (n_values[-1] * math.log2(n_values[-1])) if n_values[-1] > 1 else 1
else:
    scaling_factor = 1

log_linear_theory = [scaling_factor * n * math.log2(n) for n in n_values if n > 1]

log_linear_n_for_small_n = [scaling_factor * n * math.log2(n) if n > 1 else scaling_factor * n * 1 for n in n_values]

plt.plot(n_values, log_linear_n_for_small_n, linestyle='--', color='r', label='Theoretical O(n log n)')

plt.xlabel("Input size n")
plt.ylabel("Time (milliseconds)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort)")
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
import matplotlib.pyplot as plt
import math
import time

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0
    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return steps
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps

def log_linear_example(n):
    data_list = list(range(n))
    total_steps = 0

    for i in range(n):

        total_steps += binary_search(data_list, n - 1)
    return total_steps

n_values = []
time_values = []

for i in range(1, 10):
    n = 10 * (2**i)
    n_values.append(n)
    steps_taken = log_linear_example(n)
    time_values.append(steps_taken)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='b', label='Observed Operations (n*log n)')


if len(n_values) > 0 and n_values[-1] > 1:

    scaling_factor = time_values[-1] / (n_values[-1] * math.log2(n_values[-1]))
else:
    scaling_factor = 1

log_linear_theory = [scaling_factor * n * math.log2(n) for n in n_values]

plt.plot(n_values, log_linear_theory, linestyle='--', color='r', label='Theoretical O(n log n)')

plt.xlabel("Input size n")
plt.ylabel("Number of Operations")
plt.title("O(n log n) Log-Linear Time Complexity (n Binary Searches)")
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7: Take a suitable example and write a function that runs in polynomial time (O(n^c), e.g., O(n^2)). Plot the graph as well, similar to the above examples.
import matplotlib.pyplot as plt
import time

def polynomial_time_example(n):

    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count

n_values = []
time_values = []

for i in range(1, 15):
    n = 2**i
    n_values.append(n)


    steps_taken = polynomial_time_example(n)
    time_values.append(steps_taken)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='b', label='Observed Operations (n^2)')


if len(n_values) > 0 and n_values[-1] > 0:

    scaling_factor = time_values[-1] / (n_values[-1]**2)
else:
    scaling_factor = 1

polynomial_theory = [scaling_factor * (n**2) for n in n_values]

plt.plot(n_values, polynomial_theory, linestyle='--', color='r', label='Theoretical O(n^2)')

plt.xlabel("Input size n")
plt.ylabel("Number of Operations")
plt.title("O(n^2) Polynomial Time Complexity (Nested Loops)")
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
import time

def fibonacci_recursive(n):

    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

def measure_fibonacci_steps(n):

    memo = {}
    def _fib(k):
        if k in memo: return memo[k]
        if k <= 1:
            memo[k] = 1
            return 1
        memo[k] = 1 + _fib(k-1) + _fib(k-2)
        return memo[k]
    return 2**n

n_values = []
time_values = []

for n in range(1, 20):
    n_values.append(n)
    steps = measure_fibonacci_steps(n)
    time_values.append(steps)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='b', label='Observed Operations (2^n)')
exponential_theory = [2**val for val in n_values]

plt.plot(n_values, exponential_theory, linestyle='--', color='r', label='Theoretical O(2^n)')

plt.xlabel("Input size n")
plt.ylabel("Number of Operations")
plt.title("O(2^n) Exponential Time Complexity (Recursive Fibonacci Calls)")
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend()
plt.yscale('log')
plt.show()
```

## Quick Sort (Average Case O(n log n))

Quick Sort is a highly efficient, comparison-based sorting algorithm. It works by selecting a 'pivot' element from the array and partitioning the other elements into two sub-arrays, according to whether they are less than or greater than the pivot. The sub-arrays are then sorted recursively. The average-case time complexity of Quick Sort is **O(n log n)**, but in the worst case (when the pivot selection is consistently poor), it can degrade to O(n^2).

```python
import matplotlib.pyplot as plt
import math
import time
import random

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

n_values = []
time_values = []

for i in range(1, 10):
    n = 100 * (2**i) # Increasing input size exponentially
    n_values.append(n)

    # Generate a random list of size n. This is important for Quick Sort's average case.
    data = [random.randint(0, n) for _ in range(n)]

    start_time = time.time()
    quick_sort(data)
    end_time = time.time()

    time_values.append((end_time - start_time) * 1000) # Convert to milliseconds

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='b', label='Observed Time (ms)')

# Theoretical O(n log n) curve
if len(n_values) > 0 and n_values[-1] > 1:
    scaling_factor = time_values[-1] / (n_values[-1] * math.log2(n_values[-1])) if n_values[-1] > 1 else 1
else:
    scaling_factor = 1

log_linear_theory = [scaling_factor * n * math.log2(n) for n in n_values if n > 1]
log_linear_n_for_small_n = [scaling_factor * n * math.log2(n) if n > 1 else scaling_factor * n * 1 for n in n_values]

plt.plot(n_values, log_linear_n_for_small_n, linestyle='--', color='r', label='Theoretical O(n log n)')

plt.xlabel("Input size n")
plt.ylabel("Time (milliseconds)")
plt.title("O(n log n) Average-Case Time Complexity (Quick Sort)")
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend()
plt.xscale('log') # Use logarithmic scale for x-axis to better visualize larger n
plt.yscale('log') # Use logarithmic scale for y-axis
plt.show()
```
